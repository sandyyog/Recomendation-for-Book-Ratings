```python
# Loading Libraries and Datasets
import pandas as pd
import numpy as np
df=pd.read_csv('BX-Book-Ratings.csv')
df1=pd.read_csv('BX-Books.csv')
print(df.columns,'\n',df1.columns)
```

```python
# Merging datasets by common variable and removing unwanted variables
book=pd.merge(df,df1,on='isbn')
book.drop(['book_author','year_of_publication','publisher'],axis=1,inplace=True)
book
```

```python
# Groupby the rating by book_title
books=book.groupby('book_title')['rating'].count()
books=books.reset_index()
books=books.loc[books['rating']>10].sort_values('rating',ascending=False)
books1=pd.merge(books,book,on='book_title')
books1
```

```python
books2=books1.groupby('book_title')['rating_y'].apply(list).to_dict()
print(books2)
```

```python
# Converting books2 into Dataframe
books3=pd.DataFrame({key:pd.Series(value) for key,value in books2.items()}).fillna(0)
books3
```

```python
# Mapping Df
def encode(y):
    if y<=0:
        return 0
    if y>=1:
        return 1
final_co=books3.applymap(encode)
```

```python
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
```

```python
# Fitting into recommendation rules
freq_items=apriori(final_co,min_support=0.05,use_colnames=True)
pattern=association_rules(freq_items,metric='lift',min_threshold=0.7)
# Getting best recommendation system
pattern.sort_values('lift',ascending=False)
```
